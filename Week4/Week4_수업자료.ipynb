# Week4. 데이터 타입의 이해 - 컬렉션형

> 저번 시간에 이어 오늘은 데이터 타입 중 **컬렉션형**에 대해서 배웁니다. 기본형이 하나의 값(숫자, 문자열 등)만 담는 반면, 컬렉션형은 **여러 개의 값을 한꺼번에 담을 수 있는 그릇**입니다. 리스트, 튜플, 세트, 딕셔너리와 같은 컬렉션형은 우리가 다루는 데이터의 **집합적 성질**을 표현하는 데에 매우 유용합니다. 예를 들어, "한 사람의 나이"는 정수(int)로 표현할 수 있지만, "여행지 목록", "관광객이 방문한 활동 내역", "관광지별 평점"과 같은 데이터는 여러 값을 함께 담아야 의미가 있습니다. 이때 바로 컬렉션형이 필요합니다.  
> 이번 시간에는 각 컬렉션형의 특징을 배우고, 어떤 상황에서 어떤 타입을 선택해야 하는지 살펴보겠습니다.

## 컬렉션형

### 1. 순서형 (Sequence)

> 순서형 자료형은 여러 데이터를 **순서대로 나열**할 수 있는 타입입니다. 각 원소에는 인덱스(index)가 부여되며, 인덱싱/슬라이싱을 통해 접근할 수 있습니다.  
> 이들은 공통적으로 "순서가 있다"라는 특징을 가지지만, **수정 가능 여부(가변/불변)**, **생성 방식**에 따라 차이가 있습니다.  
> 순서형에는 **리스트(list)**, **튜플(tuple)**, **레인지(range)** 가 있습니다.

#### 리스트 (list)

- 대괄호 `[]` 를 사용하여 생성  
- 다양한 타입(숫자, 문자열 등)을 하나의 리스트에 포함 가능  
- **가변(mutable)**: 원소 추가, 삭제, 변경 가능  

```python
# 리스트 생성 및 수정
fruits = ["apple", "banana", "cherry"]
print(fruits[0])       # "apple"

fruits.append("orange")  # 원소 추가
print(fruits)          # ["apple", "banana", "cherry", "orange"]
```


#### 튜플 (tuple)
- 소괄호 () 를 사용하여 생성
- 리스트와 거의 유사하나, 불변(immutable): 원소 수정/삭제 불가
- 원소가 1개일 때는 반드시 ,를 붙여야 튜플로 인식

```python
# 튜플 생성 및 특징
data = (1, 2, 3)
print(data[1])   # 2

single = (5,)
print(type(single))   # <class 'tuple'>
```
> 튜플은 **수정되지 않는 데이터 묶음을 표현**할 때 사용합니다.</br>
> 예) 위도·경도 좌표, RGB 색상 값처럼 변하지 않는 데이터.
#### 레인지 (range)
- range(시작, 끝, 간격) 형태로 생성
- 시작 값 포함, 끝 값은 포함하지 않음
- 주로 반복문에서 활용
```python
# 레인지 생성
r = range(0, 10, 2)   # 0부터 10 전까지, 2 간격
print(list(r))        # [0, 2, 4, 6, 8]
```



#### 주의⚠️
1. 튜플(tuple)은 불변(immutable) 자료형이기 때문에, 한 번 정의된 튜플 안의 원소를 직접 바꿀 수는 없습니다.</br>
**하지만**, 리스트로 변환 후 수정하거나, 기존 튜플을 잘라서(concat/slice) 새로운 튜플을 만드는 방식으로 “수정된 효과”를 낼 수 있습니다.

##### 수정 가능한 예시1. 리스트로 변환 후 수정
```python
# 원래 튜플
t = (1, 2, 3)
print(t)        # (1, 2, 3)

# 리스트로 변환
temp = list(t)
temp[0] = 100   # 첫 번째 원소 수정
print(temp)     # [100, 2, 3]

# 다시 튜플로 변환
t = tuple(temp)
print(t)        # (100, 2, 3)
```

##### 수정가능한 예시2. 슬라이싱과 결합으로 새로운 튜플 생성
```python
t = (1, 2, 3, 4)

# 첫 번째 원소를 100으로 "수정한 것처럼" 새로운 튜플 만들기
new_t = (100,) + t[1:]
print(new_t)    # (100, 2, 3, 4)

# 중간 값 바꾸기
new_t2 = t[:2] + (99,) + t[3:]
print(new_t2)   # (1, 2, 99, 4)
```

#### ❓ 차원이란 무엇일까?

> 데이터에서 "차원"이란, 값이 배열(array)이나 집합 안에서 몇 겹으로 중첩되어 있는지를 의미합니다.  
> 1차원은 단순히 값들이 한 줄로 나열된 형태이고, 2차원은 행(row)과 열(column)로 구성된 표 형태이며, 그 이상은 고차원으로 확장됩니다.  
> 차원을 이해하는 것은 데이터를 꺼내고 가공할 때 **인덱싱과 슬라이싱**을 올바르게 활용하는 데에 중요한 역할을 합니다.

##### 🔹 차원의 구분
- **0차원 (스칼라, Scalar)**: 하나의 값(일반적으로 알고 있는 형태)
```python
x = 5
print(x)        # 5
print(type(x))  # <class 'int'>
```

- **1차원 (벡터, Vector)**: 일렬로 나열된 값들의 모음
```python
arr = [10, 20, 30]
print(arr)         # [10, 20, 30]
```
- **2차원 (행렬, Matrix)**: 행과 열로 구성된 값들의 모음
```python
matrix = [[1, 2], [3, 4]]
print(matrix)      # [[1, 2], [3, 4]]
```
- **3차원 이상 (텐서, Tensor)**: 여러 개의 행렬이 쌓인 구조
```python
tensor = [[[1, 2], [3, 4]],
          [[5, 6], [7, 8]]]
print(tensor)      # [[[1, 2], [3, 4]], [[5, 6], [7, 8]]]
```
##### 🔹 사용 예시

| 차원 | 용어 | 구조 | 예시 (데이터 분석/AI 활용) |
|------|------|------|----------------------------|
| 0차원 | 스칼라 (Scalar) | 단일 값 | 상품 가격 `100`, 도시 평균 기온 `25` |
| 1차원 | 벡터 (Vector) | 값들이 일렬로 나열 | 한 상품의 월별 매출 `[120, 130, 140, ...]` |
| 2차원 | 행렬 (Matrix) | 행 × 열 | 관광지별 월별 방문자 수 테이블, 엑셀 시트 |
| 3차원 | 텐서 (Tensor) | 여러 2차원이 층으로 쌓인 구조 | 이미지 데이터 (가로 × 세로 × 색상 채널), 연도 × 월 × 지표별 관광 통계 |
| 4차원 이상 | 고차원 텐서 | 더 복잡한 차원 | 동영상 데이터 (프레임 × 가로 × 세로 × 색상), 딥러닝 신경망 가중치 데이터 |


##### 🔹 고차원에서 인덱싱과 슬라이싱
- 인덱싱과 슬라이싱의 조합으로 특정 위치의 값을 꺼낼 수 있음
- 첫 번째 차원부터 순서대로 원소를 인덱싱, 슬라이싱할 수 있습니다.

```python
# 2차원 리스트 (행렬)
matrix = [[1, 2],
          [3, 4]]

# matrix[0] → [1, 2] (첫 번째 행)
# matrix[0][1] → 2   (첫 번째 행의 두 번째 열)
print(matrix[0][1])  # 2

# 3차원 리스트 (텐서)
tensor = [
          [ [1, 2], [3, 4] ],   # 첫 번째 층
          [ [5, 6], [7, 8] ]    # 두 번째 층
         ]

# tensor[0] → [[1, 2], [3, 4]] (첫 번째 층)
# tensor[0][1] → [3, 4]        (첫 번째 층의 두 번째 행)
# tensor[0][1][:] → [3, 4]     (슬라이싱으로 행 전체)
print(tensor[0][1][:]) # [3, 4]

"""
✅ 가시적으로 이해하기
1차원: 숫자 줄 (예: [1, 2, 3])
2차원: 표 (행/열) → 엑셀 시트
3차원: 여러 장의 표를 겹쳐둔 것 (층/행/열)
"""
```

````python
#@title 리스트(list)

# 리스트는 대괄호 [] 로 생성, 가변(mutable) → 수정 가능
fruits = ["apple", "banana", "cherry"]
print(fruits)           # ['apple', 'banana', 'cherry']
print(type(fruits))     # <class 'list'>

# 리스트 인덱싱 : 원하는 위치의 원소 추출
print(fruits[0])        # 'apple'

# 리스트에 원소 추가하기
fruits.append("orange")
print(fruits)           # ['apple', 'banana', 'cherry', 'orange']

# 인덱싱으로 원하는 위치 원소 바꾸기
fruits[1] = "blueberry" # 두 번째 원소를 blueberry로 교체
print(fruits)           # ['apple', 'blueberry', 'cherry', 'orange']

# 리스트 슬라이싱 : 원하는 범위의 원소 추출
print(fruits[1:3])      # ['blueberry', 'cherry']

# 슬라이싱으로 원하는 범위 원소 바꾸기
fruits[2:4] = ['banana', 'pear'] # 세 번째에서 네 번째 원소 교체
print(fruits)
````

#### 주의⚠️
> 인덱싱과 슬라이싱을 활용해 원소를 교체하는 경우, **교체하고자 하는 원소의 개수**를 반드시 맞춰줘야 합니다.  
> 단, **인덱싱은 단일 원소만 교체 가능**하며, 여러 개 원소를 한 번에 바꾸려면 **슬라이싱**을 사용해야 합니다.  

```python
# ✅ 인덱싱 교체 (정상)
nums = [1, 2, 3, 4]
nums[1] = 99           # 두 번째 요소만 교체
print(nums)            # [1, 99, 3, 4]

# ❌ 인덱싱으로 여러 개 교체 시도 → 리스트 자체가 원소로 추가됨
nums = [1, 2, 3, 4]
# nums[1] = [7, 8]     # 두 번째 자리에 리스트 자체가 들어감
# print(nums)          # [1, [7, 8], 3, 4] (원소 자체가 리스트로 들어감)

# ✅ 슬라이싱 교체 (개수 맞춤)
nums = [1, 2, 3, 4]
nums[1:3] = [7, 8]     # 2개 구간 → 2개 값 교체
print(nums)            # [1, 7, 8, 4]

# ⚠️ 슬라이싱 교체 (개수 불일치)
nums = [1, 2, 3, 4]
nums[1:3] = [99]       # 2개 구간 → 1개 값 교체
print(nums)            # [1, 99, 4]  (원소 개수가 줄어듦)

nums = [1, 2, 3, 4]
nums[1:3] = [9, 8, 7]  # 2개 구간 → 3개 값 교체
print(nums)            # [1, 9, 8, 7, 4]  (원소 개수가 늘어남)

````python
#@title 튜플(tuple)

# 튜플은 소괄호 () 로 생성, 불변(immutable) → 수정 불가
data = (1, 2, 3, 4)
print(data)            # (1, 2, 3, 4)
print(type(data))      # <class 'tuple'>

# 원소 접근
print(data[2])         # 3

# ⚠️ 원소 수정 시도 → 에러 발생
# data[0] = 10  # TypeError: 'tuple' object does not support item assignment

# 원소가 1개일 경우 반드시 콤마 필요
single = (5,)
print(type(single))    # <class 'tuple'>
````

````python
#@title 레인지(range)

# range(start, stop, step) → 시작 포함, 끝은 포함되지 않음
r1 = range(5)          # 0 ~ 4
print(list(r1))        # [0, 1, 2, 3, 4]

r2 = range(2, 10, 2)   # 2부터 10 전까지, 2 간격
print(list(r2))        # [2, 4, 6, 8]

print(type(r1))        # <class 'range'>
````

### 2. 집합형



> 집합형 자료형은 **중복을 허용하지 않고, 순서가 없는(unique & unordered)** 데이터들의 모임입니다.  
> 집합 이론을 그대로 구현한 타입으로, 교집합・합집합・차집합 등 수학적 연산을 직관적으로 수행할 수 있습니다.  
> 파이썬에서는 **`set`** 타입이 이에 해당합니다.
> '집합'과 같은 개념이라고 이해하시면 쉽습니다.

#### 집합 (set)

- 중괄호 `{}` 또는 `set()` 함수를 사용하여 생성  
- **순서가 없음(unordered)** → 인덱싱/슬라이싱 불가  
- **중복 자동 제거**  
- **가변(mutable)**: 원소 추가, 삭제 가능  

```python
# set 생성
numbers = {1, 2, 3, 3, 4}
print(numbers)    # {1, 2, 3, 4} → 중복 3은 자동 제거

empty = set()     # 빈 집합 생성 ( {} 는 딕셔너리 dict )
print(type(empty))  # <class 'set'>

#### 주의⚠️
1. 집합(set)은 **순서가 없는(unordered)** 자료형입니다.  
   따라서, 리스트/튜플처럼 **인덱싱(indexing)이나 슬라이싱(slicing)** 으로 원소에 접근할 수 없습니다.  
   만약 특정 원소를 꺼내고 싶다면 반복문(`for`)이나 집합 연산을 활용해야 합니다.

```python
s = {10, 20, 30}
# print(s[0])   # ❌ 오류 발생: TypeError (인덱싱 불가)

for val in s:
    print(val)  # ✅ 반복문으로 원소 꺼내기
```


2.	빈 중괄호 {} 는 빈 집합(set)이 아닌 빈 딕셔너리(dict) 로 인식됩니다.</br>
따라서, **빈 집합을 만들려면 반드시 set()을 사용**해야 합니다.
```python
a = {}

````python
#@title 집합(set)

# 집합은 중괄호 {} 또는 set() 으로 생성, 순서 없음 / 중복 허용 안함
fruits = {"apple", "banana", "cherry"}
print(fruits)           # {'apple', 'banana', 'cherry'}
print(type(fruits))     # <class 'set'>

# 집합에 원소 추가하기
fruits.add("orange")
print(fruits)           # {'apple', 'banana', 'cherry', 'orange'}

# 집합에서 원소 삭제하기
fruits.remove("banana")
print(fruits)           # {'apple', 'cherry', 'orange'}

# 안전하게 삭제 (해당 원소가 없어도 오류 발생 X)
fruits.discard("pineapple")
print(fruits)           # {'apple', 'cherry', 'orange'}

# 집합 연산: 합집합 (|), 교집합 (&), 차집합 (-), 대칭차집합 (^)
set_a = {"apple", "banana", "cherry"}
set_b = {"banana", "pear", "kiwi"}

# 합집합 → {'apple', 'banana', 'cherry', 'pear', 'kiwi'}
print(set_a | set_b)
print(set_a.union(set_b))

# 교집합 → {'banana'}
print(set_a & set_b)
print(set_a.intersection(set_b))

# 차집합 → {'apple', 'cherry'}
print(set_a - set_b)
print(set_a.difference(set_b))
````

### 3. 매핑형

> 매핑형은 **키(key)와 값(value)을 한 쌍으로 묶어서 저장하는 자료형**입니다.  
> 파이썬에서 매핑형을 대표하는 타입은 **딕셔너리(dict)**입니다.  
> 딕셔너리는 "키 : 값" 형태로 데이터를 저장하고, 키를 이용해 빠르게 값을 검색할 수 있습니다.

#### 딕셔너리 (dict)
- 중괄호 `{}`를 사용하여 생성
- 키(key)는 **변경 불가능(immutable)한 자료형**만 가능 (문자열, 숫자, 튜플 등)
- 값(value)은 모든 자료형 가능
- **순서가 보장**되며(파이썬 3.7+), 키는 중복될 수 없음

```python
# 딕셔너리 생성
person = {"name": "홍길동", "age": 30, "job": "developer"}
print(person)              # {'name': '홍길동', 'age': 30, 'job': 'developer'}
print(type(person))        # <class 'dict'>

# 키를 통해 값 조회
print(person["name"])      # "홍길동"

# 새로운 키-값 쌍 추가
person["city"] = "Seoul"
print(person)

# 기존 값 수정
person["age"] = 31
print(person)

# 키 삭제
del person["job"]
print(person)

#### 주의⚠️
1. **존재하지 않는 키에 접근하면 오류 발생**
```python
person = {"name": "홍길동", "age": 30}
print(person["height"])   # ❌ KeyError
```
2.	안전하게 접근하려면 .get() 메서드 사용
```python
print(person.get("height"))       # None
print(person.get("height", 170))  # 170 (기본값 설정)
```
3. 딕셔너리는 반복문에서 기본적으로 키를 기준으로 순회
```python
person = {"name": "홍길동", "age": 30, "city": "Seoul"}
for key in person:
    print(key, ":", person[key])
# 출력:
# name : 홍길동
# age : 30
# city : Seoul
```

4. 키, 값, 키-값 쌍을 따로 가져올 수 있음
```python
print(person.keys())    # dict_keys(['name', 'age', 'city'])
print(person.values())  # dict_values(['홍길동', 30, 'Seoul'])
print(person.items())   # dict_items([('name', '홍길동'), ('age', 30), ('city', 'Seoul')])
```

#### 반복문

````python
#@title 딕셔너리(dict)

# 딕셔너리는 중괄호 {} 안에 "키: 값" 형태로 정의, 순서 없음 / 키는 중복 불가
person = {"name": "Alice", "age": 25, "city": "Seoul"}
print(person)              # {'name': 'Alice', 'age': 25, 'city': 'Seoul'}
print(type(person))        # <class 'dict'>

# 특정 키에 접근 (인덱싱 대신 키로 접근)
print(person["name"])      # "Alice"

# 키-값 쌍 추가
person["job"] = "Engineer"
print(person)              # {'name': 'Alice', 'age': 25, 'city': 'Seoul', 'job': 'Engineer'}

# 값 수정
person["age"] = 30
print(person)              # {'name': 'Alice', 'age': 30, 'city': 'Seoul', 'job': 'Engineer'}

# 키 삭제
del person["city"]
print(person)              # {'name': 'Alice', 'age': 30, 'job': 'Engineer'}

# 안전하게 키 접근 (get) → 없는 키를 조회해도 오류 발생하지 않음
print(person.get("city"))        # None
print(person.get("city", "N/A")) # "N/A"

# 딕셔너리 주요 메서드
print(person.keys())    # dict_keys(['name', 'age', 'job'])
print(person.values())  # dict_values(['Alice', 30, 'Engineer'])
print(person.items())   # dict_items([('name', 'Alice'), ('age', 30), ('job', 'Engineer')])
````
