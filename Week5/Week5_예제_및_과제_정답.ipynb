# 예제

```python
#@title 예제 1.1. for문으로 리스트의 합을 구해봅시다.

# 지시사항:
# - for문을 사용해 numbers의 합을 구해 total에 저장하세요.
# - 빈칸(_____)을 채우세요.

numbers = [3, 5, 7, 9]
total = 0

for n in numbers:      # numbers를 순회
    total += n   # n을 누적

print(total)
```

```python
#@title 예제 1.2. 인덱스와 값을 함께 출력해봅시다.

# 지시사항:
# - 인덱스와 값을 함께 출력하는 함수를 생각해봅시다.
# - 빈칸(_____)을 채우세요.

fruits = ["apple", "banana", "cherry"]

for idx, name in enumerate(fruits, start=1):
    print(idx, name)
```

```python
#@title 예제 2.1. while문으로 1부터 N까지 합을 구해봅시다.

# 지시사항:
# - while문을 사용해 1부터 N까지의 합을 구해 total에 저장하세요.
# - 빈칸(_____)을 채우세요.

N = 5
total = 0
i = 1

while i <= N:   # i가 N 이하인 동안 반복
    total += i      # i를 누적
    i += 1          # i를 1 증가

print(total)
```

```python
#@title 예제 2.2. 무한 루프 구조에서 반복을 제어해봅시다.

# 지시사항:
# - 무한 루프를 만들기 위해서 어떤 조건을 넣어야할까요?
# - 빈칸(_____)을 채우세요.

limit = 3
count = 0

while True:            # 항상 참인 조건
    count += 1
    print("반복:", count)
    if count == limit:  # limit에 도달하면
        break           # 반복 종료
```
