> 저희가 앞으로 배우게 될 '데이터'는 다양한 타입이 존재합니다. 데이터 타입을 이해하는 것은 특정값의 연산부터 시작해서 데이터를 수집하고 결합, 수정, 삭제 등의 데이터 전처리와 추론, 예측을 위한 데이터 분석 방법론을 결정하는데에도 아주 중요한 역할을 합니다. 이번 시간을 통해서 데이터에는 어떤 타입이 있는지, 각 타입들은 어떤 식으로 다룰 수 있는지 학습해봅시다.

## 기본형

### 1. 숫자형

> 숫자형에는 일반적으로 '정수(Integer)', '실수(float)'가 있습니다. 정수와 실수의 구분은 추후 연산과 데이터 필터링을 할 때, 결과에 영향을 줄 수 있기 때문에, 숫자형 타입의 데이터 안에서도 확인하는 절차를 거쳐야합니다. type 함수로 타입을 확인하고, int(), float() 함수로 원하는 타입으로 변경도 해봅시다.

#### 정수형 (int)

	• 소수점 없는 모든 정수
	• 크기 제한 없음 (아주 큰 수, 음수도 표현 가능)

```python
# 정수로 변환하는 함수
x = 3.14
int(x) # 결과물 : 3
```
> **정수로 변환하게 되면, 정수는 소수점이 없는 숫자형이기 때문에, 소수점 밑으로는 버림이 되어 출력이 됩니다.** 추후 버림이 필요한 상황에서 이런 특성을 활용해볼 수 있겠죠?

#### 실수형 (float)

	• 소수점이 있는 수
	• 지수 표현 가능 (예: 1.23e3 → 1230.0)
	• 부동소수점 오차에 주의해야 함

```python
# 정수로 변환하는 함수
x = 10
float(x) # 결과물 : 10.0
```

#### 주의⚠️
1. 아주 큰 정수를 float 타입으로 변환하게 되면 정밀도 손실이 발생할 수 있습니다.
  * 정밀도 손실(Precision Loss) : 컴퓨터가 숫자를 표현하거나 계산할 때, 원래의 값과 약간 차이가 생기는 현상
```python
print(0.1 + 0.2)       # 0.30000000000000004
print(0.3 == 0.1+0.2)  # False

big = 12345678901234567890
print(float(big))   # 1.2345678901234568e+19 (정밀도 일부 손실)
```

2. 문자열 타입도 숫자형으로 변환이 가능하나, 숫자 형태여야만 가능합니다.(단, 앞뒤 공백은 가능, 중간에 있는 공백은 불가능)

##### ⭕️ 가능 예시
```python
print(int("12"))         # 12  (정상)
print(int(" 12 "))       # 12  (앞뒤 공백 무시)

print(float("12.5"))     # 12.5 (정상)
print(float(" 12.5 "))   # 12.5 (앞뒤 공백 무시)
```
##### ❌ 불가능 예시
```python
print(int("12.5"))     # ❌ 오류: ValueError (소수 문자열은 int로 변환 불가)

print(int("1 2"))      # ❌ 오류: ValueError (중간 공백 → 하나의 숫자가 아님)
print(float("1 2"))    # ❌ 오류: ValueError

print(int("apple"))    # ❌ 오류: ValueError (숫자 형태가 아님)
```

3. 숫자로 된 문자열 타입에도 소수점으로 되어있으면 반드시 float 함수로 변환해야 합니다.


```python
print(float(" 12.5 "))  # 12.5
print(int(" 12.5 "))    # ❌ 오류 (ValueError)
```

#### ❓ 그럼 정수와 실수는 왜 구분해야하는 것일까?

> 숫자형을 공부하다 보면 문득 이런 의문이 들 수도 있을 것입니다. 하지만, 이를 구분하는 것은 각 변수의 특성을 이해하고, 통계 분석이나 데이터 분석을 할 때 실수를 줄여줍니다.

다음 예시를 확인해봅시다.
이 표는 어느 마트의 매출표라고 합시다.

| 상품명     | 판매 수량(개) | 단가 (\$) | 매출액 (\$) |
|------------|-----------------|----------------|--------------------------|
| 사과       | 120             | 1.50           | 180.00                   |
| 바나나     | 85              | 0.98           | 83.30                    |
| 오렌지     | 200             | 1.20           | 240.00                   |
| 수박       | 15              | 9.50           | 142.50                   |

> 이 표에서 나온 여러 데이터 값은 데이터의 생김새를 보고도 파악할 수 있지만, 변수의 이름을 보고도 알 수가 있습니다. '판매 수량(개)' 컬럼(열)을 확인해보면 단위가 '개' 임을 알 수 있습니다. **우리는 일반적으로 수량을 셀 때, 120.5개라고 세지 않습니다. 따라서, 소수점이 나오면 어색한 데이터가 되는 것이죠.** 결론적으로, '판매 수량' 컬럼은 데이터를 보지 않아도 정수임을 알 수 있는 겁니다.
>
> 반대의 경우도 마찬가지입니다. '단가(\$)' 컬럼도 마찬가지로 숫자 단위가 정수로 딱 떨어지지 않음을 알 수 있습니다. 만약, 이 컬럼의 데이터 타입을 정수로 설정하게 되면 어떻게 될까요? '1.50' 값은 '1'로 표현될 것이고, '0.98'은 '0'으로 표기될 것입니다. '매출액' 컬럼도 마찬가지로 일부 값이 소실되거나, '단가' 컬럼의 잘못된 데이터 타입 설정으로 잘못된 결과를 도출할 수도 있을 것입니다.
>
> 이처럼 같은 숫자형으로 보이지만, 정수냐, 실수냐에 따라 데이터의 특성을 이해하거나 연산을 할 때, 다른 결과를 불러올 수 있습니다.

#### 문자형의 특징


````python
#@title 1. 순서(sequence) → 인덱싱, 슬라이싱 가능

word = "Python"
print(word[0])    # 'P'
print(word[:3])   # 'Pyt'
````

#### ❓ 인덱싱, 슬라이싱이란?

- 인덱싱 : 위치를 지정해서 한 글자(또는 요소)를 꺼내는 것 -> x[0]

- 슬라이싱 : 범위를 지정해서 여러 글자를 잘라내는 것 -> x[0:2]
 1) 시작 위치 포함, 끝 위치는 포함되지 않음(중요!) -> x[0:2] : 첫 번째 요소부터 두 번째 요소까지 출력
 2) 시작이나 끝을 생략할 수도 있음

##### ⚠️ **파이썬에서 첫 번째를 나타내는 숫자는 0이고, 마지막을 나타내는 숫자는 -1 입니다.**

```python
s = "Python"
print(s[0])     # 'P'
print(s[-1])    # 'n'
print(s[0:3])   # 'Pyt'
print(s[:])     # 'Python' : 전체 출력(0이나 -1은 슬라이싱할 때 생략할 수 있음)
```

````python
#@title 2. 불변(immutable) → 한 번 생성된 문자열은 수정 불가

s = "Hello"
s[0] = "h"   # ❌ 오류 발생
````

#### ❓ 포매팅(Formatting)이란?
•	포매팅 : 문자열 안에 변수나 값을 삽입해서 완성된 문장을 만드는 것
1. format() 메서드를 활용하는 방식
2. f-string 을 활용하는 방식
3. % 연산자를 활용하는 방식


##### ⚠️ 문자열을 + 연산자로 이어 붙일 수도 있지만, 포매팅을 쓰면 훨씬 간단하고 가독성이 좋아집니다. 다양한 포맷팅 메서드를 활용해서 가독성을 높혀봅시다.

```python
city = "Jeju"
visitors = 15200000
rating = 4.8

# format() 방식
print("{}는 연간 {}명이 방문합니다.".format(city, visitors))

# f-string 방식
print(f"{city}는 관광 평점이 {rating}점입니다.")

# % 방식 (구식)
print("%s는 연간 %d명이 방문합니다." % (city, visitors))
```

#### 문자열의 변환

```python
# 문자열 타입으로 변환하는 함수
num = 123
str(num)       # "123"

flag = True
str(flag)      # "True"
```
#### 주의⚠️
> **거의 모든 객체의 타입을 문자열로 변환**할 수 있습니다. 하지만, 변환은 어디까지나 읽기 좋은 문자열 표현일 뿐, 변환된 문자열을 다시 원래 타입으로 되돌릴 수 있는 것은 아닙니다.


```python
print(str(123))          # "123"
print(str(3.14))         # "3.14"
print(str(True))         # "True"
print(str([1, 2, 3]))    # "[1, 2, 3]"
print(str({"a": 1}))     # "{'a': 1}"
```

| 원래 타입 | 원래 값 | `str()` 변환 결과 | 다시 변환 가능 여부 |
|-----------|---------|-------------------|---------------------|
| 정수(int) | 123 | `"123"` | ✅ `int("123")` 로 복원 가능 |
| 실수(float) | 3.14 | `"3.14"` | ✅ `float("3.14")` 로 복원 가능 |
| 불리안(bool) | True | `"True"` | ❌ `"True"` 를 `bool()`로 직접 변환 불가 (`bool("True")` → True이지만, 모든 문자열이 True 처리됨) |
| 리스트(list) | [1, 2, 3] | `"[1, 2, 3]"` | ❌ 문자열을 그대로 `list()`로 변환 불가 |
| 딕셔너리(dict) | {"a": 1} | `"{'a': 1}"` | ❌ 문자열을 그대로 `dict()`로 변환 불가 |
| 함수(function) | `def hi(): ...` | `"<function hi at 0x...>"` | ❌ 코드 실행 없이 복원 불가 |

#### ❓ 그럼 문자열은 왜 불변이라고 하는 것일까?

> 문자열을 공부하다 보면 “여러 메서드를 통해 바꿀 수 있는데 왜 불변이라고 할까?” 라는 의문이 생길 수 있습니다. 불변이라는 것은 단순히 “바꿀 수 없다”라는 뜻이 아니라, **이미 만들어진 문자열 자체를 수정할 수는 없고, 수정한 것처럼 보이는 메서드는 사실 새로운 문자열을 생성해서 반환**한다는 의미입니다.

아래 예시를 확인해봅시다.



```python
s = "Hello"
print(s.upper())   # "HELLO"
print(s)           # "Hello"
```

> 이 코드에서 s.upper()는 기존 문자열 s를 바꾼 것이 아니라, 대문자로 변환된 **새로운 문자열 “HELLO”**를 반환한 것입니다. 원래의 s는 여전히 “Hello” 그대로 남아 있음을 확인할 수 있습니다.

```python
s = "Hello"
# s[0] = "h"    # ❌ 오류 발생 (문자열은 항목 대입 불가)
```


> 리스트와 달리 문자열의 특정 요소를 직접 바꾸려고 하면 오류가 발생합니다. 이는 문자열 객체가 불변(immutable)하기 때문입니다. **즉, 한 번 만들어진 문자열은 내부를 바꿀 수 없고, 새로운 문자열을 만들어서 교체하는 방식으로만 “변경”이 가능합니다.**
